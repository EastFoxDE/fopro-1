```python
import numpy as np
import random as rand
import matplotlib.pyplot as plt

def in_circle(x,y):
    z = np.sqrt(x*x + y*y)
    if z <= 1 :
        return 1
    return 0

def create(n):
    cCircle = n
    cCumulative = 0
    rand.seed()
    for i in range(0,n):
        x = rand.uniform(-1.0, 1.0)
        y = rand.uniform(-1.0, 1.0)
        x_values.append(x)
        y_values.append(y)
        cCumulative += in_circle(x, y)
    return cCumulative

# program start
n = 100

x_values = []
y_values = []

pi = 4*create(n) / n
print("Pi is ~", pi, ", right?")

circle = plt.Circle((0,0), 1, color='b', alpha=0.2)
plt.gca().add_patch(circle)
plt.scatter(x_values, y_values)
plt.grid()
plt.show()
```
