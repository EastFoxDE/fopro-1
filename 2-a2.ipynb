```python
import random as rand
import math
import matplotlib.pyplot as plt
rand.seed()

def f_x(x):
    return (4 / (1+x**2))

def p_x(x):
    return 1.4427 / (x+1)
    
def mc_integration(n):
    sum = 0.0
    for i in range(0, n):
        z = rand.uniform(0,1)
        x = math.e ** (z/1.4427) -1
        sum += f_x(x) / p_x(x)
    return sum / n

    
def gleich(n):
    sum = 0.0
    for i in range(0, n):
        x = rand.uniform(0,1) # NUR FÜR GLEICHVERTEILUNG
        sum += f_x(x)
    return sum / n

n = 1000
xes = []
yes = []
y2 = []

for x in range(0, n):
    xes.append(x)
    yes.append(mc_integration(n))
    y2.append(gleich(n))
               
plt.plot(xes,yes, label="f(x)")
plt.plot(xes,y2, alpha=0.3, color='b', label="g(x)")
plt.legend()
plt.grid()    
```
