```python
from diffType import *

#start
def gradient(x):
    g = 0.01             #stepsize
    p = diffType(x, 1.0) #our var to diff
    eps = 1e-7           #to stop diff
    count = 0            #for reference
    
    while(1):
        count += 1
        y = ((p**3)+(1 + p**2).sqrt())**2
        if(abs(y.dvalue) < eps):
            break
        p = p - g * y.dvalue

    print(f"Found a minimum at Point p({p.value:.3f}, {y.value:.3f}) on iteration {count}")
    
gradient(-0.5)
gradient(0)
gradient(-0.25)
```
